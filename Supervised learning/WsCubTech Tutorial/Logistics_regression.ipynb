# **LOGISTICS REGRESSION**


## **Based On Single Input**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('linear_separable_data.csv')
df.head()
```

```python
sns.pairplot(data=df)
plt.show()
```

```python
x = df[['CGPA']]
y = df['Placed']
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=14)
lr.fit(x_train,y_train)
lr.score(x_test,y_test)*100
```

```python
lr.predict([[8.796]])
```

```python
lr.predict([[5.431]])
```

```python
sns.scatterplot(x='CGPA',y='Placed',data=df)
sns.lineplot(x='CGPA',y=lr.predict(x),data=df,color="red")
plt.show()
```

## **Based On Multiple Inputs**

```python
df = pd.read_csv('collegedata.csv')
df.head()
```

```python
sns.pairplot(data=df,hue="Placed")
plt.show()
```

```python
x = df.iloc[:,:-1]
y = df['Placed']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=35)
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[4.45 , 77.32]])
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=lr)
plt.show()
```

```python
print(lr.coef_)
print(lr.intercept_)
```
