```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

```python
df = pd.read_csv('collegedata.csv')
df.head()
```

```python
df.isnull().sum()    
```

```python
sns.scatterplot(x='CGPA',y='Score',data=df,hue='Placed')
plt.show()
```

```python
x = df.iloc[:,:-1]
y = df['Placed']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=35)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
gnb = GaussianNB()
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100
```

```python
gnb.score(x_train,y_train)*100
```

```python
mlt = MultinomialNB()
mlt.fit(x_train,y_train)
```

```python
mlt.score(x_test,y_test)*100
```

```python
mlt.score(x_train,y_train)*100
```

```python
brl = BernoulliNB()
brl.fit(x_train,y_train)
```

```python
brl.score(x_test,y_test)*100
```

```python
brl.score(x_train,y_train)*100
```
