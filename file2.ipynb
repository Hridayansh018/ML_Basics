## **Feature Scaling**
**Standardization:-**
- **X_new = (X_i - X_mean)/Standard Devision**

**Normalization:-**
- **(X_i - X_min)/(X_max - X_min)**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('loan-test.csv')
dataset .head()
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset['ApplicantIncome'])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
ss.fit(dataset[['ApplicantIncome']])
```

```python
dataset['ApplicantIncome_ss']=pd.DataFrame(ss.transform(dataset[['ApplicantIncome']]),columns=['X'])
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(15,7))
plt.subplot(1,2,1)
plt.title('Before')
sns.distplot(dataset['ApplicantIncome'])

plt.subplot(1,2,2)
plt.title('After')
sns.distplot(dataset['ApplicantIncome_ss'])
plt.show()
```

- **MIN MAX SCALING**

```python
from sklearn.preprocessing import MinMaxScaler
ms = MinMaxScaler()
```

```python
ms.fit(dataset[['CoapplicantIncome']])
```

```python
dataset['CoapplicantIncome_ms']= ms.transform(dataset[['CoapplicantIncome']])
```

```python
dataset.head()
```

```python
plt.figure(figsize=(15,7))
plt.subplot(1,2,1)
plt.title('Before')
sns.distplot(dataset['CoapplicantIncome'])

plt.subplot(1,2,2)
plt.title('After')
sns.distplot(dataset['CoapplicantIncome_ms'])
plt.show()
```
